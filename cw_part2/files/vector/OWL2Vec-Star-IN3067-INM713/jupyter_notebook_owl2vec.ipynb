## Loading embeddings and getting similarities

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load word embeddings from a text file
def load_embeddings(file_path, min_vector_length=20):
    embeddings = {}
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f, 1):
            parts = line.strip().split()
            if len(parts) < min_vector_length + 1:
                print(f"Skipping line {i} (too short or missing vector): {parts[0] if parts else 'EMPTY'}")
                continue
            word = parts[0]
            try:
                vector = np.array([float(x) for x in parts[1:]])
                embeddings[word] = vector
            except ValueError:
                print(f"Skipping line {i} (non-numeric content): {word}")
    return embeddings

# Compute cosine similarity between two vectors
def compute_cosine_similarity(v1, v2):
    return cosine_similarity(v1.reshape(1, -1), v2.reshape(1, -1))[0][0]

# Analyze pairs of entities for similarity
def run_analysis(embedding_file, similar_pairs, dissimilar_pairs, threshold=0.5):
    print(f"\nLoading embeddings from: {embedding_file}")
    embeddings = load_embeddings(embedding_file)
    print(f"Loaded {len(embeddings)} vectors.\n")

    # Display the first 30 available terms to help verify content
    print("First 30 available terms in the embeddings:")
    for i, key in enumerate(embeddings.keys()):
        print(f"  {i+1}: {key}")
        if i == 29:
            break
    print()

    # Analyze and report similarity results for each pair
    def analyze(pairs, expected="Similar"):
        print(f"--- {expected} Pairs Analysis ---")
        for e1, e2 in pairs:
            if e1 not in embeddings or e2 not in embeddings:
                print(f"Missing: {e1 if e1 not in embeddings else ''} {e2 if e2 not in embeddings else ''}")
                continue
            sim = compute_cosine_similarity(embeddings[e1], embeddings[e2])
            flag = ""
            if expected == "Dissimilar" and sim > threshold:
                flag = "Too similar!"
            elif expected == "Similar" and sim < threshold:
                flag = "Not similar enough!"
            print(f"Similarity({e1}, {e2}) = {sim:.4f} {flag}")
        print()

    analyze(similar_pairs, expected="Similar")
    analyze(dissimilar_pairs, expected="Dissimilar")

# Path to the OWL2Vec* embeddings file
embedding_file = "./cache1/output/CityWatch_config1.embeddings.txt"

# Define entity pairs expected to be similar
similar_pairs = [
    ("accident", "crash"),
    ("injuries", "injury"),
    ("trafficaccident", "http://www.city.ac.uk/inm713-in3067/2025/CityWatch#TrafficAccident")
]

# Define entity pairs expected to be dissimilar
dissimilar_pairs = [
    ("injury", "intersection"),
    ("type", "light"),
    ("accident", "$1,500")
]

# Run the analysis
if __name__ == "__main__":
    run_analysis(embedding_file, similar_pairs, dissimilar_pairs)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import rdflib
from tqdm import tqdm

# Load embeddings from OWL2Vec output
def load_embeddings(filepath):
    embeddings = {}
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) < 10:  
                continue
            if parts[0].startswith("http"):
                uri = parts[0]
                try:
                    vec = np.array([float(x) for x in parts[1:]])
                    embeddings[uri] = vec
                except ValueError:
                    continue 
    return embeddings


# Enhanced class URI filter
def is_class_uri(uri):
    if not uri.startswith("http"):
        return False
    local = uri.split("#")[-1] if "#" in uri else uri.split("/")[-1]
    if "_" in local or any(char.isdigit() for char in local):
        return False
    return local and local[0].isupper()

# Generate Top-K alignments above a threshold with debug print
def generate_topk_alignments(emb1, emb2, threshold=0.3, top_k=3):
    alignments = []
    emb2_keys = list(emb2.keys())
    emb2_vectors = np.array([emb2[k] for k in emb2_keys])

    for uri1, vec1 in tqdm(emb1.items(), desc=f"Aligning (Top-{top_k})..."):
        similarities = cosine_similarity([vec1], emb2_vectors)[0]
        top_indices = similarities.argsort()[-top_k:][::-1]

        matched = False
        for idx in top_indices:
            sim = similarities[idx]
            uri2 = emb2_keys[idx]
            if sim >= threshold:
                alignments.append((uri1, uri2, sim))
                print(f"Matched: {uri1}  {uri2} (sim={sim:.4f})")
                matched = True
        if not matched:
            best_match_uri = emb2_keys[top_indices[0]]
            best_sim = similarities[top_indices[0]]
            print(f"No match for: {uri1} (best: {best_match_uri}, sim={best_sim:.4f})")

    if not alignments:
        print("No alignments found above the threshold. Try lowering it or inspect embeddings.")
    return alignments

# Save alignments to TTL file
def save_alignments_to_ttl(alignments, output_file):
    g = rdflib.Graph()
    OWL = rdflib.Namespace("http://www.w3.org/2002/07/owl#")

    for uri1, uri2, sim in alignments:
        g.add((rdflib.URIRef(uri1), OWL.equivalentClass, rdflib.URIRef(uri2)))

    g.serialize(destination=output_file, format='turtle')
    print(f"Saved {len(alignments)} alignments to {output_file}")

# Main execution
if __name__ == "__main__":
    emb1 = load_embeddings("./cache2/output/CityWatch_config2.embeddings.txt")
    emb2 = load_embeddings("./cache_lecture/output/CityWatch_config2_lecuture.embeddings.txt")

    # Filter to class-like URIs only
    emb1 = {k: v for k, v in emb1.items() if is_class_uri(k)}
    emb2 = {k: v for k, v in emb2.items() if is_class_uri(k)}

    print(f"[INFO] Filtered to {len(emb1)} CityWatch class vectors, {len(emb2)} Lecturer class vectors")

    alignments = generate_topk_alignments(emb1, emb2, threshold=0.6, top_k=3)
    save_alignments_to_ttl(alignments, r"C:\City\SWT_KG\coursework_part2\SWT_KG_CityWatch\cw_part2\files\output\vector3_alignment.ttl")
```

```python
import rdflib
from sklearn.metrics import precision_score, recall_score, f1_score

# Load triples from a TTL alignment file
def load_alignments(file_path):
    g = rdflib.Graph()
    g.parse(file_path, format="turtle")
    
    alignments = set()  # Use a set to avoid duplicate entries
    for subj, pred, obj in g:
        if pred == rdflib.URIRef("http://www.w3.org/2002/07/owl#equivalentClass"):
            alignments.add((str(subj), str(obj)))
    return alignments

# Compute evaluation metrics
def evaluate_alignment(predictions, reference):
    # Use set operations to compute TP, FP, FN
    true_positives = len(predictions.intersection(reference))
    false_positives = len(predictions - reference)
    false_negatives = len(reference - predictions)
    
    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    
    return precision, recall, f1


if __name__ == "__main__":
    # Replace with your actual file paths
    predicted_alignments = load_alignments(r"C:\City\SWT_KG\coursework_part2\SWT_KG_CityWatch\cw_part2\files\output\vector3_alignment.ttl")
    reference_alignments = load_alignments(r"C:\City\SWT_KG\coursework_part2\SWT_KG_CityWatch\cw_part2\files\reference_alignment.ttl")

    # Evaluate the predicted alignment against the reference
    precision, recall, f1 = evaluate_alignment(predicted_alignments, reference_alignments)

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

```
